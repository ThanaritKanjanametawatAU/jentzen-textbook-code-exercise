## 1. Model Definition

```python
import torch
from torch import nn
```

```python
class CNN146model:
    def __init__(self):
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=2, kernel_size=(2,2))
        self.ReLU = nn.ReLU()
        self.conv2 = nn.Conv2d(in_channels=2, out_channels=1, kernel_size=(1,1))

        with torch.no_grad():
            self.conv1.weight.set_(
                torch.Tensor([[[[0, 0], [0, 0]]], [[[1, 0], [0, 1]]]]) 
                # shape: [2, 1, 2, 2]
                # 2 out channels  ( number of kernel )
                # 1 in channels  ( kernel's channel/depth )
                # 2 kernel height
                # 2 kernel width 

                # 2 kernels, each with 1 channel
            )

            self.conv1.bias.set_(
                torch.Tensor([1, -1])
                # shape: [2]
                # 2 out channels
            )

            self.conv2.weight.set_(
                torch.Tensor([[[[-2]], [[2]]]])
                # shape: [1, 2, 1, 1]

                # 1 kernel, each with 2 channels
                # the
            )

            self.conv2.bias.set_(
                torch.Tensor([3])
                # shape: [1]

            )
            

        
    def __call__(self, x):
        print(f"x input shape: {x.shape}")

        # x1
        x = self.conv1(x)
        print(f"conv1 output shape: {x.shape}")
        x = self.ReLU(x)
        print(f"ReLU output shape: {x.shape}")

        # x2
        x = self.conv2(x)
        print(f"conv2 output shape: {x.shape}")
        self.out = x
        return self.out
```

## 2. Model Inference

```python
model = CNN146model()
```

```python
print(f"{model.conv1.weight.shape=} | {model.conv1.weight.dtype=}")
print(f"{model.conv1.bias.shape=} | {model.conv1.bias.dtype=}")
print(f"{model.conv2.weight.shape=} | {model.conv2.weight.dtype=}")
print(f"{model.conv2.bias.shape=} | {model.conv2.bias.dtype=}")
```
