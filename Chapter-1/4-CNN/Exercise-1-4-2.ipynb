```python
import torch
from torch import nn
```

```python
class CNN142Model:
    def __init__(self):
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=3, kernel_size=(1,3))
        self.ReLU = nn.ReLU()
        self.conv2 = nn.Conv2d(in_channels=3, out_channels=1, kernel_size=(1,2))

        with torch.no_grad():
            
            self.conv1.weight.set_(
                torch.Tensor([[[[1, 1, 1]]], [[[2, -2, -2]]], [[[-3, -3, 3]]]])
            )

            self.conv1.bias.set_(
                torch.Tensor([3, -2, -1])
            )

            
            self.conv2.weight.set_(
                torch.Tensor([[[[2, -1]], [[-1, 2]], [[-1, 0]]]])
            )

            self.conv2.bias.set_(
                torch.Tensor([-2])
            )




    def __call__(self, x):
        x = self.conv1(x)
        x = self.ReLU(x)
        x = self.conv2(x)
        self.out = x
        return self.out
```

```python
input = torch.Tensor([1, -1, 1, -1, 1, -1, 1, -1, 1]).unsqueeze(0).unsqueeze(0)
model = CNN142Model()
output = model(input)
output
```
